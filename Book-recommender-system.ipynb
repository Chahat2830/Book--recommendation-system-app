```python
new_book = ratings.merge(book,on= 'ISBN')
```

```python
new_book.head()
```

```python
new_ratings= new_book.groupby('Book-Title').count()['Book-Rating'].reset_index()
new_ratings.rename(columns={'Book-Rating': 'num_ratings'},inplace= True)
```

```python
new_ratings
```

```python
avg_ratings= new_book.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings.rename(columns={'Book-Rating': 'avg_ratings'},inplace= True)
```

```python
popular_df= new_ratings.merge(avg_ratings,on='Book-Title')
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings',ascending=False).head(50)
print(popular_df)
```

```python
x= new_book.groupby('User-ID').count()['Book-Rating']>200
educated_users= x[x].index
```

```python
filtred_rating=new_book[new_book['User-ID'].isin(educated_users)]
```

```python
y=filtred_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_rating=filtred_rating[filtred_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.drop_duplicates()
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt = pt.fillna(0)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
score = cosine_similarity(pt)
```

```python
score
```

```python
def recommed(book_name):
    # Check if book_name exists in pt.index
    if book_name not in pt.index:
        return f"Book '{book_name}' not found in dataset."

    index = np.where(pt.index == book_name)[0][0]
    score_items = sorted(list(enumerate(score[index])), key=lambda x: x[1], reverse=True)[1:6]

    suggestions = []
    for i in score_items:
        suggestions.append(pt.index[i[0]])

    return suggestions
```

```python
recommended_books = recommed('1984')
print(recommended_books)
```
